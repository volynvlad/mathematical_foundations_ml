```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import itertools
```

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

print(quicksort([3,6,8,10,1,2,1]))
```

1.1 Пусть в переменных x и y заданы некоторые наборы случайных данных. Построить и обучить модель линейной регрессии. Сделать выводы о точности модели.

```python
rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y);
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
reg = LinearRegression().fit(x, y)
```

```python
reg.score(x, y)
```

```python
coef = reg.coef_[0][0]
inter = reg.intercept_[0]
coef, inter
```

```python
plt.scatter(x, y);
plt.plot([0, 10], [inter, 10 * coef + inter], color="#FF0000", label="predicted")
plt.plot([0, 10], [-1, 10 * 2 - 1], color="#00FFFF", label="actual")
plt.legend()
plt.show()
```

```python
def ordinary_least_square(X, y):
    X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
#     inv = np.linalg.inv(np.dot(np.transpose(X - np.mean(X)), (X - np.mean(X)))) 
#     X_y = np.dot(np.transpose(X - np.mean(X)), y)
    inv = np.linalg.inv(np.dot(np.transpose(X), (X))) 
    X_y = np.dot(np.transpose(X), y)
    k = np.dot(inv, X_y)
    def predict(X):
        X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
        return X @ k
    return predict
```

```python
from sklearn.mixture import GaussianMixture
```

```python
gm = GaussianMixture(n_components=3, covariance_type="full", random_state=42).fit(pca_X_train)
```

```python
gm.predict(pca_X_test)
```

```python
def map(name):
    if name == 'setosa':
        return 0
    elif name == 'virginica':
        return 1
    else:
        return 2
```

```python
pca_y_test = gm.predict(pca_X_test)
```

```python
import matplotlib as mpl
```

```python
# plot_results(X_test, y_test, gm.means_, gm.covariances_, 0, "Gaussian Mixture")

color_iter = itertools.cycle(["red", "blue", "green"])
splot = plt.subplot(2, 1, 1)
for i, (mean, covar, color) in enumerate(zip(gm.means_, gm.covariances_, color_iter)):
    v, w = np.linalg.eigh(covar)
    v = 2.0 * np.sqrt(2.0) * np.sqrt(v)
    u = w[0] / np.linalg.norm(w[0])
    if not np.any(y_test == i):
        continue
    print(f"i = {i}")
    plt.scatter(pca_X_test[pca_y_test == i], pca_X_test[pca_y_test == i], 0.8, color=color)

    angle = np.arctan(u[1] / u[0])
    angle = 180.0 * angle / np.pi
    ell = mpl.patches.Ellipse(mean, v[0], v[1], 180.0 + angle, color=color)
    ell.set_clip_box(splot.bbox)
    ell.set_alpha(0.5)
    splot.add_artist(ell)

plt.xlim(-9.0, 5.0)
plt.ylim(-3.0, 6.0)
plt.xticks(())
plt.yticks(())
plt.title("Gausian Mixture")

plt.show()
```
